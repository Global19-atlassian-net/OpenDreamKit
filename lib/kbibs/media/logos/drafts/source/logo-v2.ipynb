```python
%matplotlib inline

import pylab
pylab.xkcd(scale=2)
```

```python
import numpy as np
```

## Basic idea 

- some parametrise curve
- plotted with the human touch (exploiting the xkcd features, see http://matplotlib.org/xkcd/examples/showcase/xkcd.html)
- keep it simple

```python
n = 3
t = np.linspace(0, 2*np.pi*n, 1000)

a = -1
b = 1
x = a* 0.5 * np.sin(2 * t) + b * np.cos(t/n)
y = a* 0.4 * np.cos(2 * t) + b * np.sin(t/n)
pylab.plot(x,y,'-', linewidth=5, color='green')
pylab.axis('off')
```

## Use fewer periods in the independent variable t -- looks like a person:

```python
n = 2
t = np.linspace(0, 2*np.pi*n, 1000)

a = -1
b = 1
x = a* 0.5 * np.sin(2 * t) + b * np.cos(t/n)
y = a* 0.4 * np.cos(2 * t) + b * np.sin(t/n)
pylab.plot(x,-y,'-', linewidth=5, color='green')
pylab.axis('auto')
pylab.axis('off')
pylab.savefig('logo-icon.pdf')
pylab.savefig('logo-icon.png')
pylab.savefig('logo-icon.svg')
```

## Adding project name below icon (unsquashed)

```python
n = 2
t = np.linspace(0, 2*np.pi*n, 1000)

a = -1
b = 1
x = a* 0.5 * np.sin(2 * t) + b * np.cos(t/n)
y = a* 0.4 * np.cos(2 * t) + b * np.sin(t/n)
pylab.figure(figsize=(5, 5))
pylab.plot(x,-y,'-', linewidth=5, color='green')
pylab.text(-1.5,-1.7,'OpenDreamKit',fontsize=40, color='green')
pylab.axis('off')
# need to set margins manually as text is not recognised by 'auto'
pylab.axis([-1.5, 1.7, -1.7, 1.5])
pylab.savefig('logo1.pdf')
pylab.savefig('logo1.png')
pylab.savefig('logo1.svg')
pylab.axis()
```

## With project name; a little bit 'squashed' looks more human:

```python
n = 2
t = np.linspace(0, 2*np.pi*n, 1000)

a = -1
b = 1
x = a* 0.5 * np.sin(2 * t) + b * np.cos(t/n)
y = a* 0.4 * np.cos(2 * t) + b * np.sin(t/n)
pylab.figure(figsize=(5, 5))
pylab.plot(x*1.1,-y/1.2,'-', linewidth=5, color='green')
pylab.text(-1.68,-1.7,'OpenDreamKit',fontsize=40, color='green')
pylab.axis('off')
#pylab.axis('equal')
# need to set margins manually as text is not recognised by 'auto'
pylab.axis([-1.7, 1.7, -1.7, 1.5])
pylab.savefig('logo2.pdf')
pylab.savefig('logo2.png')
pylab.savefig('logo2.svg')
pylab.axis()
```
